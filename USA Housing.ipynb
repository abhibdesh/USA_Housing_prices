# Setting X and y variables to the arrays

```python
X=housing[['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms','Avg. Area Number of Bedrooms','Area Population']]
y=housing['Price']
```

# Importing Scikitlearn and splitting the dataset in train and test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

# Importing Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

# Creating and training the model

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
accuracy6 = lm.score(X_test,y_test)
print("Accuracy: {:.2f} %".format(accuracy6.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracy6.std()*100))
```

# Model Evaluation

```python
print(lm.intercept_)
```

```python
coeff_df=pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
p=lm.predict(X_test)
```

```python
l=lm.predict(X)
m=housing['Price']
plt.figure(figsize=(13,7))
plt.scatter(l,m)
plt.show()
```

```python
sns.displot((y_test-p),bins=50)
```
